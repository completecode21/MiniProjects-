```python
import pandas as pd
df = pd.read_csv('parkinsons_du.csv', index_col=None)
df.head()
```

### a. identifying and differentiating the dataset in Numerical and Categorical Data &nbsp; <br> <br> b. Categoriacal Data Representation- countplot, piechart, etc., <br> <br>c. Numerical Data Representation- histogram, distplot, boxplot, etc.,

```python
#only one categoical column is present in dataset status
import seaborn as sns
import matplotlib.pyplot as plt
```

## Categorical data 

```python
sns.countplot(df,x='status')
df['status'].value_counts()
#1-parkinson
#0-healthy
```

```python
df['status'].value_counts().plot(kind='pie',autopct='%1.0f%%')
```
